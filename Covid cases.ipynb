## aggragate the data for covid cases 

```python
import numpy as np
import pandas as pd
```

### read confirmed cases

#### Data preparation 

##### remove colums , switching rows with colums(transposed) and adding the total colum so sum the row

```python
## read the csv
confirmed_df = pd.read_csv("confirmed_cases_global.csv", index_col=False)
## to avoind the indexed colunm we set this as the the indexed 
confirmed_df.set_index('Country/Region', inplace=True)
confirmed_df= confirmed_df.drop(["Lat","Long","Province/State"], axis=1)
confirmed_df = confirmed_df.transpose()
confirmed_df["Total confirmed"] = confirmed_df.sum(axis = 1, skipna = True, numeric_only=True) 

confirmed_df.head(3)
confirmed_df.tail(3)
```


### Read dead cases


#### Data preparation 


##### remove colums , switching rows with colums(transposed) and adding the total colum so sum the row.

```python
death_df = pd.read_csv("deaths_cases_global.csv", index_col=False)
death_df.set_index('Country/Region', inplace=True)
death_df= death_df.drop(["Lat","Long","Province/State"], axis=1)
death_df = death_df.transpose()
death_df["Total death"] = death_df.sum(axis = 1, skipna = True, numeric_only=True) 

death_df.head(3)
```

#### Extract from each table the "total" and merge the two "total" columns into one .

```python
agg_data = pd.concat([death_df["Total death"],  confirmed_df["Total confirmed"]], axis=1)
agg_data
```

### Export to csv do it can be used on next step for the stocks. 

```python
agg_data.to_csv('aggregated_data.csv', index=True )
```
